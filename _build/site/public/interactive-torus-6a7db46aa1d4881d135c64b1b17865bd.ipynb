```python
N = round(r*np.pi/h)
r=r.value
R=R.value
h=h.value
# Stitch count
st_count = [0]*(N+1)
row = [0]*(N+1)
for n in range(N+1):
    theta = n*np.pi/N
    st_count[n] = str(round(2*np.pi*(R-r*np.cos(theta))/w))
    row[n] = n #"Row "+str(n)+": "+str(st_count[n])+" st."
```
