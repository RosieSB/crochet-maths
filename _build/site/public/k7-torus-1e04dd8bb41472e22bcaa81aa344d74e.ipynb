# 7 colouring of a torus

```python
from tabulate import tabulate

import numpy as np

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from matplotlib.colors import TABLEAU_COLORS, same_color
%matplotlib ipympl

R = 8
r = 3
w = .6
h = .6

# Pattern instructions
N = round(r * np.pi/h)
st_count = [0]*(int(N)+1)
pattern = [0]*(int(N)+1)
row = [0]*(int(N)+1)
for n in range(int(N)+1):
    st_count[n]=round(2*np.pi*(R-r*np.cos(n*np.pi/int(N)))/w)
    row[n] = n #"Row "+str(n)+": "+str(st_count[n])+" st."
```

```python
plt.clf()

# Pattern model
fig = plt.figure(figsize = (10,10),label=' ')
ax = plt.axes(projection='3d')
ax.grid()

colours = [0]*(N+1)

for n in range(int(N)+1):
    colours[n] = ['yellow' for i in range(st_count[n])]
    for k in range(0,st_count[n]):
        theta = n*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k)*2*np.pi/st_count[n], (k+1)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        m=-.5
        if 0<=phi<np.pi/4:
            if -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'magenta'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'orange'
        elif np.pi/4<=phi<=3*np.pi/4:
            if theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'magenta'
            else:
                colours[n][k] = 'lime'
        elif 3*np.pi/4<=phi<=4*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'lime'
            elif -(-1/m)*theta-phi+4*np.pi<=theta<=-(-1/m)*theta-phi+6*np.pi:
                colours[n][k] = 'red'
        elif 4*np.pi/4<=phi<=5*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'cyan'
            elif -(-1/m)*theta-phi+4*np.pi<=theta<=-(-1/m)*theta-phi+6*np.pi:
                colours[n][k] = 'red'
        elif 5*np.pi/4<=phi<=6*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'cyan'
            elif -(-1/m)*theta-phi+4*np.pi<=theta<=-(-1/m)*theta-phi+6*np.pi:
                colours[n][k] = 'orange'
        elif 6*np.pi/4<=phi<=7*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'cyan'
            else:
                colours[n][k] = 'orange'
        elif 7*np.pi/4<phi<8*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi+2*np.pi<=theta<=-(-1/m)*theta-phi+4*np.pi:
                colours[n][k] = 'magenta'
            else:
                colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)

        theta = -n*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((-n-.5)*np.pi/N, (-n+.5)*np.pi/N, 2)
        v = np.linspace((k)*2*np.pi/st_count[n], (k+1)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        if 0<=phi<=np.pi/4:
            if -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi-4*np.pi<=theta<=-(-1/m)*theta-phi-2*np.pi:
                colours[n][k] = 'orange'
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'magenta'
        elif np.pi/4<=phi<=2*np.pi/4:
            if -(-1/m)*theta-phi<=theta<=-(-1/m)*theta-phi+2*np.pi:
                colours[n][k] = 'magenta'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi-4*np.pi<=theta<=-(-1/m)*theta-phi-2*np.pi:
                colours[n][k] = 'lime'
        elif 2*np.pi/4<=phi<=3*np.pi/4:
            if -(-1/m)*theta-phi<=theta:
                colours[n][k] = 'magenta'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'red'
            elif -(-1/m)*theta-phi-4*np.pi<=theta<=-(-1/m)*theta-phi-2*np.pi:
                colours[n][k] = 'lime'
        elif 3*np.pi/4<=phi<=4*np.pi/4:
            if -(-1/m)*theta-phi<=theta:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'red'
            elif -(-1/m)*theta-phi-4*np.pi<=theta<=-(-1/m)*theta-phi-2*np.pi:
                colours[n][k] = 'lime'
        elif 4*np.pi/4<=phi<=5*np.pi/4:
            if -(-1/m)*theta-phi<=theta:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'red'
        elif 5*np.pi/4<=phi<=6*np.pi/4:
            if -(-1/m)*theta-phi<=theta:
                colours[n][k] = 'blue'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'orange'
        elif 6*np.pi/4<=phi<8*np.pi/4:
            if -(-1/m)*theta-phi<=theta:
                colours[n][k] = 'gray'
            elif -(-1/m)*theta-phi-2*np.pi<=theta<=-(-1/m)*theta-phi:
                colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
        
# Axis labels
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

# Axis limits
ax.set(zlim=(-R-r,R+r))
    
ax.set_aspect('equal')

plt.show()
```
