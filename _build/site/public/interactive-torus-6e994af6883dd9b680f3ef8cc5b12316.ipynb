```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

R_widget = widget.FloatSlider(min=0.0, max=10.0, step=0.05)
R = widget.FloatSlider(min=0.5, max=10.0, step=0.05, value=5.0)

def update_R(*args):
    R = R_widget
R_widget.observe(update_R, 'value')

def printer(R_widget, R):
    print(R_widget, R)
interact(printer,x=R_widget, y=R);
```
