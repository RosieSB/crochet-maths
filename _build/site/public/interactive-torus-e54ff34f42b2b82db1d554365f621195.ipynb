```python
#| label: r_value

r_slide = widget.FloatSlider(min=0, max=10, step=.01, value=1.78, description='r')
r = r_slide.value
r_slide
```

```python
#| label: w_value

w_slide = widget.FloatSlider(min=0, max=2, step=.01, value=.4, description='w')
w = w_slide.value
w_slide
```

```python
#| label: h_value

h_slide = widget.FloatSlider(min=0, max=2, step=.01, value=.4, description='h')
h = h_slide.value
h_slide
```

```python
#| label: N_value

N = round(r*np.pi/h)
```

```python
#| label: Stitch_count
st_count = [0]*(N+1)
row = [0]*(N+1)
for n in range(N+1):
    theta = n*np.pi/N
    st_count[n] = str(round(2*np.pi*(R-r*np.cos(theta))/w))
    row[n] = n 
```

Parameters: major radius R = {eval}`R`; minor radius r = {eval}`r`; stitch height h = {eval}`h`; stitch width w = {eval}`w`. 

Number of rows: $N = \lfloor\frac{r\pi}{h}\rfloor = ${eval}`N`.

**Row 0.** Chain {eval}`st_count[0]` and join into round. 1 st into each chain sp around.

**Subsequent rows:** Work the following stitch counts in subsequent rows. Increases made using 2 st in st of previous row. Try to evenly distribute increases for a symmetrical look.

```python
from tabulate import tabulate
import numpy as np

table = np.transpose([row, st_count])
pattern = tabulate(table,headers=["Row","Stitch count"],tablefmt="plain")
print(pattern)
```
