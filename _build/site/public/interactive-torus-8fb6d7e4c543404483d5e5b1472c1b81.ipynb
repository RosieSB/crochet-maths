```python
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
import numpy as np
from tabulate import tabulate

R = widgets.FloatText(value=4,min=0,step=0.1,description='Major radius')
r = widgets.FloatText(value=1.8,min=0,step=0.1,description='Minor radius')
w = widgets.FloatText(value=.4,min=0,step=0.1,description='Stitch width')
h = widgets.FloatText(value=.4,min=0,step=0.1,description='Stitch height')

def f(R,r,w,h):
    N = round(r*np.pi/h)
    # Stitch count
    st_count = [0]*(N+1)
    row = [0]*(N+1)
    for n in range(N+1):
        theta = n*np.pi/N
        st_count[n] = str(round(2*np.pi*(R-r*np.cos(theta))/w))
        row[n] = n #"Row "+str(n)+": "+str(st_count[n])+" st."
    #table = np.transpose([row, st_count])
    #pattern = tabulate(table,headers=["Row","Stitch count"],tablefmt="plain")
    print('Set up: Chain ', st_count[0],'; 1 stitch into each chain sp.,\n\nSubsequent rows: Work the following stitches evenly around each subsequent row:\n\n','pattern','\n\nThis completes the upper half of the torus.\n\nFor the lower half, turn work upside down and repeat Rows 1-',N-1,'. Fill and sew up.')
    
ui = widgets.HBox([R, r, w, h])

out = widgets.interactive_output(f, {'R': R, 'r': r, 'w': w, 'h':h})

display(ui, out)
```
