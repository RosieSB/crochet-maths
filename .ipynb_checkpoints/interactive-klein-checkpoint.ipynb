```python
import ipywidgets as widgets
from ipywidgets import Layout

import numpy as np
import matplotlib.pyplot as plt

d = widgets.FloatSlider(value=-1,min=-5,max=0,step=.1,description='d')
r = widgets.FloatSlider(value=1,min=0,step=.1,description='r')
s = widgets.FloatSlider(value=2,min=0,step=.1,description='s')
theta = widgets.FloatSlider(value=1,min=0,max=np.pi/2,step=.1,description='θ')

def f(d,r,s,theta):
    plt.clf()
    plt.close()
    
    h = (r+s+(r-s)*np.cos(theta))/np.sin(theta)
    
    P = [0,r*(1+np.cos(theta)),h-r*np.sin(theta)]
    plt.scatter(P)
    
    Q = [0,s*(1-np.cos(theta)),s*np.sin(theta)]
    PQ = np.array(Q)-np.array(P)
    magPQ = np.sqrt(PQ[1]**2+PQ[2]**2)
    
    fig = plt.figure(figsize = (10,5), label = ' ')
    ax = plt.axes(projection='3d')
    
    #Straight segment 1
    t = np.linspace(0, -d, 2)
    
    x = 0
    y = 0
    z = -t
    
    ax.plot(x, y, z)
    
    #Straight segment 2
    u = np.linspace(0, np.pi, 2)
    
    t = np.linspace(d, h, 2)
    
    x = 0
    y = 0
    z = t
    
    ax.plot(x, y, z)
    
    #Circular segment 1
    
    t = np.linspace(0, r*(np.pi+theta), 100)
    
    x = 0
    y = r*(1+np.cos(np.pi-t/r))
    z = h+r*np.sin(np.pi-t/r)
    
    ax.plot(x, y, z)
    
    #Straight segment 3
    
    t = np.linspace(0, magPQ, 2)
    
    x = (t/magPQ)*Q[0] + (1-t/magPQ)*P[0]
    y = (t/magPQ)*Q[1] + (1-t/magPQ)*P[1]
    z = (t/magPQ)*Q[2] + (1-t/magPQ)*P[2]
    
    ax.plot(x, y, z)
    
    #Circular segment 2
    
    t = np.linspace(0, s*theta, 100)
    
    x = 0
    y = s+s*np.cos(np.pi-theta+t/s)
    z = s*np.sin(np.pi-theta+t/s)
    
    ax.plot(x, y, z)
    
    # Axis labels
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    
    ax.set(xlim=(-2,2), ylim=(-2*max(r,s),2*max(r,s)), zlim=(1.5*d,1.5*(h+r)))
    
    ax.set_aspect('equal')
    
    ax.view_init(elev=0, azim=0, roll=0)
    
    plt.show()

output =  widgets.interactive_output(f, {'d': d, 'r': r, 's': s, 'theta': theta})

parameters = widgets.VBox([widgets.HTML(value="<b>Choose pattern dimensions (cm):"), d, r, s, theta],layout=Layout())

widgets.HBox([parameters,output],layout=Layout())
```
