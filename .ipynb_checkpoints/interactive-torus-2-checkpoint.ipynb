```python
import ipywidgets as widgets
import numpy as np

R = widgets.FloatText(value=4,min=0,max=30,step=.01,description='Major radius')
r = widgets.FloatText(value=1.78,min=0,max=30,step=.01,description='Minor radius')
w = widgets.FloatText(value=.4,min=0,max=5,step=.01,description='Stitch width')
h = widgets.FloatText(value=.4,min=0,max=5,step=.01,description='Stitch height')

N = widgets.Label(value=f'{round(r.value * np.pi/h.value)}')
def update_N_1(n):
    N.value = f'{round(n["owner"].value * np.pi/h.value)}'
    init_st_cnt.value = f'{round(2*np.pi*(R.value-r.value)/w.value)}'
r.observe(update_N_1)

def update_N_2(m):
    N.value = f'{round(r.value * np.pi/m["owner"].value)}'
    init_st_cnt.value = f'{round(2*np.pi*(R.value-r.value)/w.value)}'
h.observe(update_N_2)

def f(R,r,w,h):    
    st_count = [0]*(int(N.value)+1)
    for n in range(int(N.value)+1):
        st_count[n]=round(2*np.pi*(R-r*np.cos(n*np.pi/int(N.value)))/w)
        if n<10:
            print('Row  ',n,': ',st_count[n],' st')
        else:
            print('Row ',n,': ',st_count[n],' st')
            
def on_value_change(k):
    out.clear_output()
    with out:
        st_count = [0]*(int(k['new'])+1)
        for n in range(int(k['new'])+1):
            st_count[n]=round(2*np.pi*(R.value-r.value*np.cos(n*np.pi/int(k['new'])))/w.value)
            if n<10:
                print('Row  ',n,': ',st_count[n],' st')
            else:
                print('Row ',n,': ',st_count[n],' st')
            
out = widgets.interactive_output(f, {'R': R, 'r': r, 'w': w, 'h': h})

widgets.HBox([widgets.VBox([R,r,w,h]), out])

N.observe(on_value_change, names='value')
```
