```python
import numpy as np

# Major & minor radii
R = 4
r = 1.78
# Stitch dimensions
w = .4
h = .4
# Row count
N = round(r*np.pi/h)
# Stitch count
st_count = [0]*(N+1)
row = [0]*(N+1)
for n in range(N+1):
    theta = n*np.pi/N
    st_count[n] = str(round(2*np.pi*(R-r*np.cos(theta))/w))
    row[n] = "Row "+str(n)+": "+str(st_count[n])+" st."
```

```python
# Source - https://stackoverflow.com/a/64290695
# Posted by sultanalyst
# Retrieved 2026-02-22, License - CC BY-SA 4.0

import ipywidgets as widgets
checkboxes = [widgets.Checkbox(value=False, description=label) for label in row]
output = widgets.VBox(children=checkboxes)
display(output)
```
