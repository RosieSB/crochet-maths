```python
%matplotlib ipympl
import matplotlib.pyplot as plt
from matplotlib import colormaps
import numpy as np
from tabulate import tabulate


# Major & minor radii
R = 4
r = 1.78
# Stitch dimensions
w = .4
h = .4
# Row count
N = round(r*np.pi/h)
# Stitch count
st_count = [0]*(N+1)
row = [0]*(N+1)
for n in range(N+1):
    theta = n*np.pi/N
    st_count[n] = round(2*np.pi*(R-r*np.cos(theta))/w)
    #print(2*np.pi*(R-r*np.cos(theta))/w-st_count[n])
    row[n] = n #"Row "+str(n)+": "+str(st_count[n])+" st."

plt.style.use('seaborn-v0_8-poster')

#PLOTTING
# Plot crocheted(?) torus
ax = plt.figure().add_subplot(projection='3d')
ax.view_init(elev=28, azim=60, roll=0)
for n in range(N+1):
    # Create a grid of the u and v values for row n
    u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 4)
    v = np.linspace(0, 2 * np.pi, st_count[n]+1)
    #print(v)
    print('Row ',n,': ',st_count[n],' st,\t len(v)=',len(v),'\t Circumference: ',2*np.pi*(R-r*np.cos(n*np.pi/N)),'\t st count * st width: ',st_count[n]*.4)
    u, v = np.meshgrid(u, v)
    x = (R - r * np.cos(u)) * np.cos(v)
    y = (R - r * np.cos(u)) * np.sin(v)
    z = r * np.sin(u)
    ax.plot_surface(x, y, z,cmap='viridis', edgecolor='black', linewidth=.1)

# Axis limits
ax.set(zlim=(-R-r,R+r))

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

plt.show()
```

```python
matplotlib.pyplot.close()
```
