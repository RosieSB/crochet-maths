```python
from tabulate import tabulate

import numpy as np

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from matplotlib.colors import TABLEAU_COLORS, same_color
%matplotlib ipympl

R = 8
r = 3
w = .6
h = .6

# Pattern instructions
N = round(r * np.pi/h)
st_count = [0]*(int(N)+1)
pattern = [0]*(int(N)+1)
row = [0]*(int(N)+1)
for n in range(int(N)+1):
    st_count[n]=round(2*np.pi*(R-r*np.cos(n*np.pi/int(N)))/w)
    row[n] = n #"Row "+str(n)+": "+str(st_count[n])+" st."
    plt.clf()

# Pattern model
fig = plt.figure(figsize = (10,10),label=' ')
ax = plt.axes(projection='3d')
ax.grid()

colours = [0]*(N+1)

# UPPER
for n in range(int(N)+1):
    colours[n] = ['ywllow' for i in range(st_count[n])]
    K = int(st_count[n]/8)
    for k in range(0,K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'magenta'
        elif L1<=n<=L2:
            colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
    
    for k in range(K,3*K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'magenta'
        elif L1<=n<=L2:
            colours[n][k] = 'lime'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)

    for k in range(3*K,4*K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'blue'
        elif L1<=n<=L2:
            colours[n][k] = 'lime'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
       
    for k in range(4*K,5*K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'blue'
        elif L1<=n<=L2:
            colours[n][k] = 'cyan'
        elif L2<=n:
            colours[n][k] = 'red'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
        
    for k in range(5*K,6*K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'blue'
        elif L1<=n<=L2:
            colours[n][k] = 'cyan'
        elif L2<=n:
            colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
       
    for k in range(6*K,7*K):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'gray'
        elif L1<=n<=L2:
            colours[n][k] = 'cyan'
        elif L2<=n:
            colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
       
    for k in range(7*K,st_count[n]):
        theta = (n+.5)*np.pi/N
        phi = (k+.5)*2*np.pi/st_count[n] # slight offset to better reflect crochet stitch distribution
        u = np.linspace((n-.5)*np.pi/N, (n+.5)*np.pi/N, 2)
        v = np.linspace((k-.5)*2*np.pi/st_count[n], (k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'gray'
        elif L1<=n<=L2:
            colours[n][k] = 'magenta'
        elif L2<=n:
            colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)

# LOWER
for n in range(1,int(N)):
    colours[n] = ['yellow' for i in range(st_count[n])]
    K = int(st_count[n]/8)
    for k in range(0,K):
        u = np.linspace((-n-.5)*np.pi/N, (-n+.5)*np.pi/N, 2)
        v = np.linspace((-k-.5)*2*np.pi/st_count[n], (-k+.5)*2*np.pi/st_count[n], 2)
        u, v = np.meshgrid(u, v)
        x = (R-r*np.cos(u))*np.cos(v)
        y = (R-r*np.cos(u))*np.sin(v)
        z = r*np.sin(u)
        # n=m*k+(2/3)*N
        m=-(2/3)*N/st_count[n]
        L1 = m*k+(2/3)*N
        L2 = m*k+(4/3)*N
        L3 = m*k+2*N
        if n<=L1:
            colours[n][k] = 'magenta'
        elif L1<=n<=L2:
            colours[n][k] = 'gray'
        elif L1<=n<=L2:
            colours[n][k] = 'orange'
        ax.plot_surface(x, y, z, color=colours[n][k], edgecolor='black', linewidth=.1)
    


# Plot curve on torus
#v = np.linspace(-np.pi, 2*np.pi, 50)
#u = -(1/3)*v+2*np.pi/3
#x = (R - r * np.cos(u)) * np.cos(v)
#y = (R - r * np.cos(u)) * np.sin(v)
#z = r * np.sin(u)
#ax.plot3D(x, y, z,linewidth=3,color='k')


# Axis labels
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

# Axis limits
ax.set(zlim=(-R-r,R+r))
    
ax.set_aspect('equal')

plt.show()
```
